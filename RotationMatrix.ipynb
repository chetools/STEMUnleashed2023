```python
import numpy as np
from plotly.subplots import make_subplots
```

```python
x=np.linspace(0,6*np.pi,100)
y=np.sin(x)
y2=np.sin(2*x+np.pi/4)
```

```python
fig=make_subplots(2,3)
fig.add_scatter(x=x,y=y, mode='markers',name='plot1',row=1,col=1)
fig.add_scatter(x=x,y=y2, mode='lines',marker_color='rgba(100,20,30,100)',name='plot2',row=2,col=3)
fig.update_layout(width=600, height=400, template='plotly_dark')
```

```python
r=2.
theta=np.linspace(0,2*np.pi,50)
x=r*np.cos(theta)+0.5
y=r*np.sin(theta)+1.0
fig=make_subplots()
fig.add_scatter(x=x,y=y,mode='markers')
fig.update_xaxes(range=[-5,5])
fig.update_yaxes(range=[-5,5])
fig.update_layout(width=500,height=500, template='plotly_dark')
```

```python
theta= np.linspace(0,2*np.pi, 36, endpoint=False)
rot_mat=np.array([[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]])
rot_mat=np.moveaxis(rot_mat,[0,1,2],[2,0,1])
v=np.array([r,0.]).reshape(2,1)
rot_vecs=rot_mat @ v
```

```python
x,y = rot_vecs.squeeze()
```

```python
fig=make_subplots()
fig.add_scatter(x=x,y=y,mode='markers')
fig.update_xaxes(range=[-5,5])
fig.update_yaxes(range=[-5,5])
fig.update_layout(width=500,height=500, template='plotly_dark')
```

```python
stretch_mat = np.array([[2,0],[0,1]])
```

```python
vs=np.stack([x,y],axis=1).reshape(-1,2,1)
```

```python
x_stretch,y_stretch= (stretch_mat[None,:,:] @ vs).squeeze().T
```

```python
fig=make_subplots()
fig.add_scatter(x=x_stretch,y=y_stretch,mode='markers')
fig.update_xaxes(range=[-5,5])
fig.update_yaxes(range=[-5,5])
fig.update_layout(width=500,height=500, template='plotly_dark')
```

```python
theta2=np.pi/6
rot_mat2=np.array([[np.cos(theta2),-np.sin(theta2)],[np.sin(theta2),np.cos(theta2)]])
```

```python
vs=np.stack([x_stretch,y_stretch],axis=1).reshape(-1,2,1)
```

```python
x_stretch_rot,y_stretch_rot= (rot_mat2[None,:,:] @ vs).squeeze().T
```

```python
fig=make_subplots()
fig.add_scatter(x=x_stretch_rot,y=y_stretch_rot,mode='markers')
fig.update_xaxes(range=[-5,5])
fig.update_yaxes(range=[-5,5])
fig.update_layout(width=500,height=500, template='plotly_dark')
```
