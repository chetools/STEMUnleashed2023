```python
import numpy as np
from scipy.optimize import minimize
from plotly.subplots import make_subplots
```

```python
rng=np.random.RandomState(123)
pc=1.
A=450
B=-190
C=-0.05
m=0.35
T=1.7
omega=9.25
tc=88
tend=0.99*tc
t = np.linspace(82,tend,100)
p = A + B*( ((tc-t)**m) * (1+ C*np.cos(omega*np.log((tc-t)/T))))
p=rng.normal(loc=p,scale=np.abs(p)*0.03)
```

```python

def model(v):
    tc, m, omega = v

    tctm = (tc-t)**m
    wlogtct = omega*np.log(tc-t)
    V = np.c_[np.ones_like(t), tctm, tctm*np.cos(wlogtct), tctm*np.sin(wlogtct)]
    x = np.linalg.solve(V.T @ V, V.T @ p)
    model = V @ x
    return model, x
```

```python
def sse(v):
    return np.sum((model(v)[0]-p)**2)
```

```python
sol=minimize(sse,[tc, m, omega])
```

```python
fig=make_subplots()
fig.add_scatter(x=t,y=p,mode='markers')
fig.add_scatter(x=t,y=model(sol.x)[0],mode='lines')
fig.update_layout(width=800,height=600,template='plotly_dark',showlegend=False)
```
